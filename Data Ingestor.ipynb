# Sentiment 140 Datasets

```python
import pandas as pd 
import numpy as np 
import urllib.request
import zipfile
```

#### Get dataset from HuggingFace website

```python
test_url='https://huggingface.co/datasets/stanfordnlp/sentiment140/resolve/refs%2Fconvert%2Fparquet/sentiment140/test/0000.parquet'

train_url='https://huggingface.co/datasets/stanfordnlp/sentiment140/resolve/refs%2Fconvert%2Fparquet/sentiment140/train/0000.parquet'

urllib.request.urlretrieve(test_url, "test.parquet")
urllib.request.urlretrieve(train_url, "train.parquet")
```

```python

df_raw = pd.read_parquet("train.parquet")
df_raw.info()
df_raw
```

```python
df_raw_test = pd.read_parquet("test.parquet")
df_raw_test.info()
df_raw_test
```

```python
# url='https://cs.stanford.edu/people/alecmgo/trainingandtestdata.zip'
# !wget {url}
# !unzip trainingandtestdata.zip

url = "https://cs.stanford.edu/people/alecmgo/trainingandtestdata.zip"
urllib.request.urlretrieve(url, "data.zip")

with zipfile.ZipFile("data.zip", "r") as zip_ref:
    zip_ref.extractall()
```
